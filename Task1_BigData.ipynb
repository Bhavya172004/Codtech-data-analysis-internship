```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
df.describe()
```

```python
# Average bill
print("Average bill:", df['total_bill'].mean())

# Male vs Female spending
print(df.groupby('sex')['total_bill'].mean())
```
