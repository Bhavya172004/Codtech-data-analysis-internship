```python
from textblob import TextBlob

texts = [
    "I love this product",
    "This is terrible",
    "Very good experience",
    "Worst service ever"
]

for text in texts:
    analysis = TextBlob(text)
    print(text, "->", analysis.sentiment)
```

```python
# Positive or Negative
for text in texts:
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        print(text, "Positive")
    else:
        print(text, "Negative")
```
